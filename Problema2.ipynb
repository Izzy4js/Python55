```python
import pandas as pd
df = pd.read_csv('./data/winemag-data-130k-v2.csv')
df.head()
```

```python
#Explore el dataframe
print("Forma del DataFrame:", df.shape)
print("Columnas:", df.columns)
print("Primeras 5 filas:\n", df.head())
print("Descripción estadística:\n", df.describe())
```

```python
#Renombre de Columnas
df.rename(columns={
    'points': 'puntuacion',
    'price': 'precio',
    'province': 'provincia',
    'description': 'descripcion',
    'designation': 'designación',
}, inplace=True)

df.columns
```

```python
#Cracion de columnas
#columna continente
continentes_df = pd.read_csv('https://gist.githubusercontent.com/kintero/7d1db891401f56256c79/raw/a61f6d0dda82c3f04d2e6e76c3870552ef6cf0c6/paises.csv')
continentes_df.rename(columns={' name': 'country'}, inplace=True)
df = pd.merge(df, continentes_df[['country', 'continente']], how='left', on='country')

# Columna función de su precio.
df['rango_precio'] = pd.cut(df['precio'], bins=[0, 20, 50, 100, 500], labels=['Bajo', 'Medio', 'Alto', 'Premium'])

# Etiquetar vinos caros (precio > 100)
df['es_caro'] = df['precio'] > 100

#Columna que categorice la puntuación en rangos
df['categoria_puntuacion'] = pd.cut(df['puntuacion'], bins=[0, 80, 85, 90, 95, 100],
                                    labels=['Malo', 'Aceptable', 'Bueno', 'Muy Bueno', 'Excelente'])

df.head()
```

```python
#Crea 4 reportes distintos
import pandas as pd
import sqlite3
from pymongo import MongoClient

# Reporte 1: Promedio de precio y cantidad de vinos por continente
reporte_1 = df.groupby('continente')['precio'].agg(['mean', 'count']).reset_index()
reporte_1.rename(columns={'mean': 'promedio_precio', 'count': 'cantidad_vinos'}, inplace=True)

# Reporte 2: Vinos mejor puntuados por país
reporte_2 = df.groupby('country')['puntuacion'].max().reset_index().sort_values(by='puntuacion', ascending=False)

# Reporte 3: Promedio de precio por categoría de puntuación
reporte_3 = df.groupby('categoria_puntuacion')['precio'].mean().reset_index()

# Reporte 4: Conteo de vinos por rango de precio
reporte_4 = df['rango_precio'].value_counts().reset_index()
reporte_4.columns = ['rango_precio', 'cantidad_vinos']

print(reporte_1.head())
print(reporte_2.head())
print(reporte_3.head())
print(reporte_4.head())
```

```python
#Exportar los reportes 
#CSV
reporte_1.to_csv('CSV-Problema2/promedio_precio_y_cantidad_vinos_por_continente.csv', index=False)
reporte_2.to_csv('CSV-Problema2/vinos_mejor_puntuados_por_pais.csv', index=False)
reporte_3.to_csv('CSV-Problema2/promedio_precio_por_pais.csv', index=False)
reporte_4.to_csv('CSV-Problema2/vinos_por_categoria_y_continente.csv', index=False)
```

```python
#EXCEL
reporte_1.to_excel('EXCEL-Problema2/reporte_promedio_precio_cantidad.xlsx', index=False)
reporte_2.to_excel('EXCEL-Problema2/reporte_vinos_mejor_puntuados.xlsx', index=False)
reporte_3.to_excel('EXCEL-Problema2/reporte_promedio_puntuacion_por_continente.xlsx', index=False)
reporte_4.to_excel('EXCEL-Problema2/reporte_vinos_por_precio.xlsx', index=False)
```

```python
import sqlite3

conn = sqlite3.connect('SQLite-Problema2/vinos_reportes.db')

reporte_1.to_sql('reporte_promedio_precio_cantidad', conn, if_exists='replace', index=False)
reporte_2.to_sql('reporte_vinos_mejor_puntuados', conn, if_exists='replace', index=False)
reporte_3.to_sql('reporte_promedio_puntuacion_por_continente', conn, if_exists='replace', index=False)
reporte_4.to_sql('reporte_vinos_por_precio', conn, if_exists='replace', index=False)

conn.close()
```

```python
from pymongo import MongoClient
import pandas as pd

client = MongoClient('mongodb://localhost:27017/')


db = client['vinos_db']

def drop_collection_if_exists(collection_name):
    if collection_name in db.list_collection_names():
        db[collection_name].drop()


drop_collection_if_exists('reporte_promedio_precio_cantidad')
db['reporte_promedio_precio_cantidad'].insert_many(reporte_1.to_dict('records'))

drop_collection_if_exists('reporte_vinos_mejor_puntuados')
db['reporte_vinos_mejor_puntuados'].insert_many(reporte_2.to_dict('records'))

drop_collection_if_exists('reporte_promedio_puntuacion_por_continente')
db['reporte_promedio_puntuacion_por_continente'].insert_many(reporte_3.to_dict('records'))

drop_collection_if_exists('reporte_vinos_por_precio')
db['reporte_vinos_por_precio'].insert_many(reporte_4.to_dict('records'))


client.close()

import os

os.makedirs('MongoDB-Problema2', exist_ok=True)


reporte_1.to_json('MongoDB-Problema2/reporte_promedio_precio_cantidad.json', orient='records', lines=True)
reporte_2.to_json('MongoDB-Problema2/reporte_vinos_mejor_puntuados.json', orient='records', lines=True)
reporte_3.to_json('MongoDB-Problema2/reporte_promedio_puntuacion_por_continente.json', orient='records', lines=True)
reporte_4.to_json('MongoDB-Problema2/reporte_vinos_por_precio.json', orient='records', lines=True)
```
